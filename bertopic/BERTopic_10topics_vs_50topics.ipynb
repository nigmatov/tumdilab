```python
from bertopic import BERTopic
import numpy as np
```

```python
topic_model_10topics = BERTopic.load("default_bertopic_top50k_reduced10")
topic_model_50topics = BERTopic.load("default_bertopic_top50k_reduced50")
```

```python
topic_model_10topics.get_topic_info()
```

```python
with open('topic_modelling_bertopic_evaluation.txt', 'w') as fout:
    fout.write("10 Topics Model:\n")
    fout.write(json.dumps(topic_model_10topics.get_topics(), indent=4))
    fout.write("\n")
```

```python
with open('topic_modelling_bertopic_evaluation.txt', 'a') as fout:
    fout.write("50 Topics Model:\n")
    fout.write(json.dumps(topic_model_50topics.get_topics(), indent=4))
    fout.write("\n")
```

```python
import json
docs = []
with open('top50kaaa', 'r') as f:
  for line in f:
      json_doc = json.loads(line)
      docs.append(json_doc['_source']['content_t'])
```

```python
import json
test_docs = []
with open('test_random_docs_topicm', 'r') as f:
  for line in f:
      json_doc = json.loads(line)
      test_docs.append(json_doc['_source']['content_t'])
```

```python
test_docs[0]
```

```python
with open('topic_modelling_bertopic_evaluation.txt', 'a') as fout:
    topics, probs = topic_model_10topics.transform(test_docs)
    idx_file = 0
    results = ""
    fout.write("#" + str(idx_file + 1) + ":\n")
    fout.write(test_docs[idx_file])

    fout.write("\n\nTopics results (10 topics model):\n")
    
    fout.write("Topic indices and probs:")
    results += '['
    for idx in range(0,10):
        results += str(idx) + ","
    results += "]  "
    
    results += '['
    for idx in range(0,10):
        results += str(probs[idx_file][idx]) + ","
    results += "]\n"

    fout.write(results)
    
    topics, probs = topic_model_50topics.transform(test_docs)
    fout.write("\n\nTopics results (50 topics model):\n")
    results = ""
    
    fout.write("Topic indices and probs:")
    results += '['
    for idx in range(0,50):
        results += str(idx) + ","
    results += "]  "
    
    results += '['
    for idx in range(0,50):
        results += str(probs[idx_file][idx]) + ","
    results += "]\n"

    fout.write(results)

    fout.write("\n")
```

```python
from bertopic import BERTopic
```

```python
topic_model_10topics = BERTopic.load("default_bertopic_top50k_reduced10")
```

```python
topic_model_10topics.get_topic_info()
```

```python
import json
docs = []
with open('segmentaaa-output_with_embeddings', 'r') as f:
  for line in f:
      json_doc = json.loads(line)
      docs.append(json_doc['_source']['content_t'])
```

```python
topics, probs = topic_model_10topics.transform(docs)
```

```python
import pickle
with open('segmentaaa_bertopic10_topics.pickle', 'wb') as fout:
    pickle.dump(probs, fout)
```

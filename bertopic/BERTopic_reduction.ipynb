```python
import json
import numpy as np

docs = []
with open('top100kaaa', 'r') as f:
  for line in f:
      json_doc = json.loads(line)
      docs.append(json_doc['_source']['content_t'])
```

```python
from bertopic import BERTopic
```

```python
topic_model = BERTopic.load("default_bertopic_top100k")
```

```python
topics, probs = topic_model.transform(docs)
new_topics, new_probs = topic_model.reduce_topics(docs, topics, probabilities=probs, nr_topics=50)
topic_model.save("default_bertopic_top100k_reduced50")
```
